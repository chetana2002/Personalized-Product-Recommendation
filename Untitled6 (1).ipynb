# 5. Building Popularity Based Recommender model

### Manual Popularity Based Model

```python
# Calculate the count of user_id for each unique product as recommendation score
train_data_grouped = train_data.groupby('productId').agg({'userId': 'count'}).reset_index()
train_data_grouped.rename(columns={'userId': 'score'}, inplace=True)
train_data_grouped.head(10)
```

```python
# Sort the products based on recommendation score
train_data_sort = train_data_grouped.sort_values(['score', 'productId'], ascending=[False, True])

# Generate a recommendation rank based on the score
train_data_sort['rank'] = train_data_sort['score'].rank(ascending=False, method='first')

# Get the top 5 recommendations
popularity_recommendations = train_data_sort.head(5)
popularity_recommendations
```

```python
# Sample function to use a manual popularity-based recommender model to make predictions
def recommend(user_id):
    user_recommendations = popularity_recommendations.copy()

    # Add user_id column for which the recommendations are being generated
    user_recommendations['userId'] = user_id

    # Bring the user_id column to the front
    cols = user_recommendations.columns.tolist()
    cols = cols[-1:] + cols[:-1]
    user_recommendations = user_recommendations[cols]

    return user_recommendations
```

```python
# Since this is a popularity-based recommender model, all three users will receive the same recommendations.
# The recommendations are generated based on popularity and are not personalized to individual users.
find_recom = [10, 100, 150]  # List of user IDs for which recommendations will be generated.

for user_id in find_recom:
    print("List of recommendations for userId: %d\n" % user_id)
    recommendations = recommend(user_id)
    print(recommendations)
    print("\n")
```

#### Evaluation of the manual popularity based model

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# Calculate the RMSE of the popularity-based recommendation system
# The ratings present in the test data are the actual ratings (Act_rating)
test_data2 = test_data.copy()
test_data2.drop(['userId'], axis=1, inplace=True)
test_data2.rename(columns={'ratings': 'Act_rating'}, inplace=True)
```

```python
# Calculate the sum of ratings for each unique product as recommendation score
train_data_grouped2 = train_data.groupby('productId').agg({'ratings': 'sum'}).reset_index()
train_data_grouped2.rename(columns={'ratings': 'Sum_rating'}, inplace=True)
train_data_grouped2.head()
```

```python
# Merge the grouped and sorted dataframes to get inner joined results
train_data_inner = pd.merge(train_data_grouped2, train_data_sort)
train_data_inner.head()
```

```python
# Calculate the average rating of the product across users
train_data_inner["Avg_Rating"] = train_data_inner["Sum_rating"] / train_data_inner["score"]
train_data_inner.head()
```

```python
# Merge the train data with average rating and the test data with actual rating at the product level
test_data_inner = pd.merge(train_data_inner, test_data2)
test_data_inner.head()
```

```python
import math

# Calculate RMSE for the popularity-based recommender system
mse = mean_squared_error(test_data_inner["Act_rating"], test_data_inner["Avg_Rating"])
rmse = math.sqrt(mse)
print("RMSE for popularity-based recommendation system:", rmse)
```

```python
from surprise import Dataset, Reader, KNNBasic
import pandas as pd

# Create a Surprise reader specifying the rating scale
reader = Reader(rating_scale=(1, 5))

# Load the data into the Surprise dataset
data = Dataset.load_from_df(train_data[['userId', 'productId', 'ratings']], reader)

# Build the full training set
trainset = data.build_full_trainset()

# Specify similarity options
similarity_options = {
    'name': 'cosine',   # Use cosine similarity
    'user_based': False  # Item-item similarity
}

# Create the KNNBasic model and fit it on the trainset
item_sim_model = KNNBasic(sim_options=similarity_options)
item_sim_model.fit(trainset)

# Define a function to get recommendations for specific users
def get_recommendations(user_id, model, k=5):
    try:
        user_inner_id = trainset.to_inner_uid(user_id)
    except ValueError:
        print(f"User {user_id} is not part of the trainset.")
        return []

    item_ids = [iid for iid in trainset.all_items() if trainset.knows_item(iid)]
    user_items = trainset.ur[user_inner_id]
    item_ratings = [(iid, model.predict(user_inner_id, iid).est) for iid in item_ids]
    recommended_items = sorted(item_ratings, key=lambda x: x[1], reverse=True)[:k]
    return recommended_items

# Specify the user IDs for which you want recommendations
user_ids = ['A2CX7LUOHB2NDG', 'A2NWSAGRHCP8N5', 'A2WNBOD3WNDNKT', 'A1GI0U4ZRJA8WN']

# Print the recommendations in the desired format
result_rows = []

for user_id in user_ids:
    recommendations = get_recommendations(user_id, item_sim_model, k=5)
    if recommendations:
        for rank, (item_id, estimated_rating) in enumerate(recommendations, start=1):
            result_rows.append([user_id, item_id, estimated_rating, rank])
    else:
        print(f"No recommendations for user {user_id}.")

if result_rows:
    # Create a Pandas DataFrame to display the result
    result_df = pd.DataFrame(result_rows, columns=['userId', 'productId', 'score', 'rank'])
    print(result_df)
```

```python
from collections import defaultdict
def get_top_n_recommendations(reccomemndations, n=5):
    # First map the reccommendations to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in reccomemndations:
        top_n[uid].append((iid, est))

    #sort predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```
